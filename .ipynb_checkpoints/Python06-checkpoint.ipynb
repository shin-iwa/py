```python
import sqlite3
```

```python
df_csv = pd.read_csv('data.csv',encoding = 'shift-jis')
df_csv
```

```python
dbname = 'TEST.db'
conn = sqlite3.connect(dbname)
conn.close()
```

```python
conn = sqlite3.connect(dbname)

df_csv.to_sql('test_table',conn,if_exists = 'replace')

conn.close()
```

```python
conn = sqlite3.connect(dbname)

df_db = pd.read_sql('SELECT * FROM test_table', conn)

conn.close()
```

```python
df_db
```

```python
sql = '''
select * 
from test_table
where 都道府県名 == "東京都"
'''
```

```python
conn = sqlite3.connect(dbname)

df_db = pd.read_sql(sql, conn)

conn.close()
```

```python
df_db
```

```python
conn = sqlite3.connect(dbname)

for i in range(1000):
    df_csv.to_sql('test_table_1000',conn,if_exists = 'append')

conn.close()
```

```python
sql = '''
select count(*) as cnt 
from test_table_1000
'''
```

```python
conn = sqlite3.connect(dbname)

df_db = pd.read_sql(sql, conn)

conn.close()
```

```python
df_db
```

```python
sql = '''
select * 
from test_table_1000
where 都道府県名 == "東京都" and 西暦（年） =="1920.0"
'''
```

```python
conn = sqlite3.connect(dbname)

df_db = pd.read_sql(sql, conn)

conn.close()
```

```python
df_db
```
