## 11 mergeメソッド

```python
import pandas as pd
```

```python
df01 = pd.DataFrame({'氏名':['高橋', '伊藤','鈴木', '佐藤'],
                    'クラス':['a001', 'a001','a001', 'a001'],
                    '数学':[1,2,3,4]})
df01
```

```python
df02 = pd.DataFrame({'氏名':['高橋', '伊藤', '鈴木', '佐藤'],
                    'クラス':['a001', 'a001', 'a001', 'a001'],
                    '国語':[5,6,7,8]})
df02
```

```python
pd.merge(df01, df02, on='氏名')
```

```python
pd.merge(df01, df02, on='氏名', suffixes=['_left', '_right'])
```

```python
pd.merge(df01, df02, on='氏名', suffixes=['', '_重要'])
```

```python
df02_name = pd.DataFrame({'名前':['高橋', '伊藤', '鈴木', '佐藤'],
                         'クラス':['a001', 'a001', 'a001', 'a001'],
                         '国語':[5,6,7,8]})
df02_name
```

```python
pd.merge(df01, df02_name, left_on='氏名', right_on='名前')
```

```python
df03 = pd.DataFrame({'名前':['高橋', '高橋'],
                    'クラス':['a001', 'a002'],
                    '英語':[15, 16]})
df03
```

```python
pd.merge(df01, df03, left_on=['氏名', 'クラス'], right_on=['名前', 'クラス'])
```

```python
df01
```

```python
df04 = pd.DataFrame({'氏名':['高橋', '伊藤', '渡辺', '加藤'],
                    'クラス':['a001', 'a001', 'a001', 'a001'],
                    '国語':[5,6,7,8]})
df04
```

```python
pd.merge(df01, df04, on='氏名', how='left')
```

```python
pd.merge(df01, df04, on='氏名', how='right')
```

```python
pd.merge(df01, df04, on='氏名', how='outer')
```

```python
pd.merge(df01, df04, on='氏名', how='inner')
```

```python
pd.merge(df01, df04, on='氏名')
```

```python
pd.merge(df01, df04, on='氏名', how='outer')
```

```python
pd.merge(df01, df04, on='氏名', how='outer', indicator=True)
```

```python
df05 = pd.DataFrame({'クラス':['a001', 'a001', 'a001', 'a001'],
                    '数学':[1,2,3,4]},
                    index=['高橋', '伊藤', '鈴木', '佐藤'])
df05
```

```python
df02
```

## 10 concat

```python
## mergeメソッドは、データフレーム同士を横に結合、　concatメソッドは、データフレーム同士を縦に結合する。
```

```python
import pandas as pd
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df01 = pd.DataFrame({'氏名':['佐藤', '鈴木', '高橋', '田中'],
                    'クラス':['df01', 'df01', 'df01', 'df01'],
                    '数学':[1,2,3,4],
                    '国語':[5,6,7,8]})
df01
```

```python
df02 = pd.DataFrame({'氏名':['佐藤', '鈴木', '山本'],
                    'クラス':['df02', 'df02', 'df02'],
                    '数学':[9,10, 11],
                    '国語':[12,13,14]})
df02
```

```python
pd.concat([df01, df02])
```

```python
pd.concat([df01, df02], ignore_index=True)
```

```python
df03 = pd.DataFrame({'氏名':['佐藤', '鈴木', '加藤'],
                    'クラス':['df03', 'df03', 'df03'],
                    '数学':[15,16,17],
                    '国語':[18,19,20]})
df03
```

```python
pd.concat([df01, df02, df03])
```

```python
df = pd.concat([df01, df02, df03], keys=['１番目', '２番目', '３番目'])
df
```

```python
df.index
```

```python
df04 = pd.DataFrame({'氏名':['佐藤', '鈴木', '佐々木'],
                    'クラス':['df04', 'df04', 'df04'],
                    '数学':[21,22,23],
                    '社会':[24,25,26]})
df04
```

```python
df01
```

```python
pd.concat([df01, df04])
```

```python
pd.concat([df01, df04], join='outer')
```

```python
pd.concat([df01, df04], join='inner')
```

```python
pd.concat([df01, df04], axis=0)
```

```python
pd.concat([df01, df04], axis=1)
```
