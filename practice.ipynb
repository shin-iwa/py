## 11 mergeメソッド

```python
import pandas as pd
```

```python
df01 = pd.DataFrame({'氏名':['高橋', '伊藤','鈴木', '佐藤'],
                    'クラス':['a001', 'a001','a001', 'a001'],
                    '数学':[1,2,3,4]})
df01
```

```python
df02 = pd.DataFrame({'氏名':['高橋', '伊藤', '鈴木', '佐藤'],
                    'クラス':['a001', 'a001', 'a001', 'a001'],
                    '国語':[5,6,7,8]})
df02
```

```python
pd.merge(df01, df02, on='氏名')
```

```python
pd.merge(df01, df02, on='氏名', suffixes=['_left', '_right'])
```

```python
pd.merge(df01, df02, on='氏名', suffixes=['', '_重要'])
```
